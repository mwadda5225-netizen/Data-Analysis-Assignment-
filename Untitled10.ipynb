```python
Y_var = 'interest_rate'
X_vars_cont_and_orig_dummy = [
    'debt_to_income', 'total_credit_utilized', 'public_record_bankrupt',
    'annual_income', 'loan_amount', 'inquiries_last_12m' # credit_checks proxy
]
X_vars_cat = ['term', 'grade', 'emp_length', 'homeownership', 'loan_purpose']

df_model5 = df.copy()
all_vars_to_check = X_vars_cont_and_orig_dummy + X_vars_cat + [Y_var]
df_model5 = df_model5.dropna(subset=all_vars_to_check)

df_model5['emp_length'] = df_model5['emp_length'].astype(str)

df_model5['term_60'] = (df_model5['term'] == 60).astype(int)

grade_dummies = pd.get_dummies(df_model5['grade'], prefix='grade', drop_first=False, dtype=int)
grade_dummies = grade_dummies.drop('grade_A', axis=1) # Reference 'A'

emp_dummies = pd.get_dummies(df_model5['emp_length'], prefix='emp', drop_first=False, dtype=int)
emp_dummies = emp_dummies.drop('emp_10.0', axis=1) # Reference '10.0'

home_dummies = pd.get_dummies(df_model5['homeownership'], prefix='home', drop_first=False, dtype=int)
home_dummies = home_dummies.drop('home_RENT', axis=1) # Reference 'RENT'

purpose_dummies = pd.get_dummies(df_model5['loan_purpose'], prefix='purpose', drop_first=False, dtype=int)
purpose_dummies = purpose_dummies.drop('purpose_debt_consolidation', axis=1) # Reference 'debt_consolidation'

X_continuous = df_model5[X_vars_cont_and_orig_dummy]
X_dummies = pd.concat([df_model5['term_60'], grade_dummies, emp_dummies, home_dummies, purpose_dummies], axis=1)
X = pd.concat([X_continuous, X_dummies], axis=1)

X = sm.add_constant(X)
Y = df_model5[Y_var]

model_5 = sm.OLS(Y, X)
results_5 = model_5.fit()

print(results_5.summary())
```

```python
predictions = results_5.predict(X)
residuals = Y - predictions

first_five_obs = df_model5.head(5).copy()
first_five_obs['Predicted_Interest_Rate'] = predictions.head(5)
first_five_obs['Residuals'] = residuals.head(5)

report_df = first_five_obs[['interest_rate', 'Predicted_Interest_Rate', 'Residuals']]

print("\nResiduals for the first five observations:")
print(report_df)

report_df.to_csv("model5_residuals.csv", index=False)
```

```python
cols_for_analysis = [
    'interest_rate', 'annual_income', 'loan_amount',
    'debt_to_income', 'term', 'homeownership'
]
df_model = df[cols_for_analysis].copy()

df_model.dropna(subset=cols_for_analysis, inplace=True)

df_model['term'] = df_model['term'].astype(str)

print("Unique values for 'term':")
print(df_model['term'].unique())
print("\nUnique values for 'homeownership':")
print(df_model['homeownership'].unique())
print("\nSample size after dropping NaNs:", len(df_model))
```

```python
# Convert 'term' back to int and then to a category for patsy to handle it correctly
# Or keep it as object/string as patsy will treat a string column as a categorical variable by default
df_model['term'] = df_model['term'].astype(str)

def run_regression_and_extract_results(formula, data, model_name):
    """Runs OLS regression and extracts key results for the table."""
    try:
        model = smf.ols(formula, data=data).fit()

        # R-squared, Sample Size, F-statistic
        r_squared = model.rsquared
        nobs = int(model.nobs)
        f_statistic = model.fvalue

        # Coefficients and standard errors
        results = {}
        for param in model.params.index:
            coef = model.params[param]
            se = model.bse[param]
            p_value = model.pvalues[param]

            # Determine significance stars
            stars = ''
            if p_value < 0.01:
                stars = '***'
            elif p_value < 0.05:
                stars = '**'
            elif p_value < 0.10:
                stars = '*'

            # Format coefficient and standard error
            formatted_coef = f'{coef:.3f}{stars}'
            formatted_se = f'({se:.3f})'

            results[param] = {
                'coef': formatted_coef,
                'se': formatted_se
            }

        return {
            'model_name': model_name,
            'results': results,
            'r_squared': f'{r_squared:.4f}',
            'nobs': nobs,
            'f_statistic': f'{f_statistic:.2f}'
        }
    except Exception as e:
        print(f"Error running model {model_name}: {e}")
        return None

# Define formulas
formulas = {
    'Model 1': 'interest_rate ~ annual_income',
    'Model 2': 'interest_rate ~ annual_income + loan_amount',
    'Model 3': 'interest_rate ~ annual_income + loan_amount + debt_to_income',
    'Model 4': 'interest_rate ~ annual_income + loan_amount + debt_to_income + C(term)',
    'Model 5': 'interest_rate ~ annual_income + loan_amount + debt_to_income + C(term) + C(homeownership)'
}

# Run models and collect results
all_results = []
for name, formula in formulas.items():
    result = run_regression_and_extract_results(formula, df_model, name)
    if result:
        all_results.append(result)

# Step 4: Table Generation

# Identify all unique parameters across all models
all_params = set()
for res in all_results:
    all_params.update(res['results'].keys())

# Sort parameters: Intercept first, then alphabetically for other variables
param_order = ['Intercept'] + sorted([p for p in all_params if p != 'Intercept'])
param_map = {
    'Intercept': 'Intercept',
    'annual_income': 'Annual Income',
    'loan_amount': 'Loan Amount',
    'debt_to_income': 'Debt-to-Income',
    'C(term)[T.60]': 'Term (60 months)',
    'C(homeownership)[T.OWN]': 'Homeownership (OWN)',
    'C(homeownership)[T.RENT]': 'Homeownership (RENT)'
    # The reference categories for C(term) and C(homeownership) are term=36 and homeownership=MORTGAGE, respectively.
}

# Create a list of rows for the table
table_rows = []
model_names = [res['model_name'] for res in all_results]

# 1. Parameter rows
for param in param_order:
    # Skip if parameter is not in the map (e.g., a reference category that patsy drops)
    if param not in param_map and param != 'Intercept':
        continue

    row_name = param_map.get(param, param)
    row_data_coef = [row_name]
    row_data_se = ['']

    for res in all_results:
        if param in res['results']:
            row_data_coef.append(res['results'][param]['coef'])
            row_data_se.append(res['results'][param]['se'])
        else:
            row_data_coef.append('') # Empty for coefficient
            row_data_se.append('') # Empty for standard error

    table_rows.append(row_data_coef)
    table_rows.append(row_data_se)

# Add separator
table_rows.append(['---'] * (len(model_names) + 1))

# 2. Goodness-of-fit rows
gof_rows = [
    ('R-squared', 'r_squared'),
    ('N', 'nobs'),
    ('F-statistic', 'f_statistic')
]

for row_name, key in gof_rows:
    row_data = [row_name] + [res[key] for res in all_results]
    table_rows.append(row_data)

# Create the final DataFrame
columns = ['Variable'] + model_names
df_table = pd.DataFrame(table_rows, columns=columns)

# Save to CSV for the user
output_filename = 'regression_results_table.csv'
df_table.to_csv(output_filename, index=False)

print(f"Regression results saved to {output_filename}")
print(df_table.to_string())
```
